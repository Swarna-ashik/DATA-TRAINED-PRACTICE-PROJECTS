# work with seaborn library

```python
import pandas as pd
import seaborn as sns
semp=pd.read_csv('salaryofemp.csv')
print(semp)
```

# Stripplot

```python
ax=sns.stripplot(x='rank',y='salary',data=semp)
plt.title('Graph')
plt.show()
```

```python
ax=sns.stripplot(x='yrs.since.phd',y='salary',data=semp)
plt.title('Graph')
plt.show()
```

```python
ax=sns.stripplot(x='yrs.service',y='salary',data=semp)
plt.title('Graph')
plt.show()
```

```python
ax=sns.stripplot(x='sex',y='salary',data=semp)
plt.title('Graph')
plt.show()
```

# Swarmplot

```python
ax=sns.swarmplot(x='rank',y='salary', data=semp)
plt.title('Graph')
plt.show()
```

```python
ax=sns.swarmplot(x='yrs.since.phd',y='salary', data=semp)
plt.title('Graph')
plt.show()
```

```python
ax=sns.swarmplot(x='yrs.service',y='salary', data=semp)
plt.title('Graph')
plt.show()
```

```python
ax=sns.swarmplot(x='sex',y='salary', data=semp)
plt.title('Graph')
plt.show()
```
